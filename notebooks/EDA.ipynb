```python
# Amharic POS Tagger: Exploratory Data Analysis

#This notebook performs EDA on the parsed Amharic .conllu dataset.
import pandas as pd
import matplotlib.pyplot as plt
# Now import your module
from src.parse_conllu import parse_conllu

```

```python
# Load sentences
sentences = parse_conllu("../data/am_att-ud-test.conllu")
```

```python
# Flatten into dataframe
rows = [(word, tag) for s in sentences for word, tag in s]
df = pd.DataFrame(rows, columns=["Word", "Tag"])
```

```python
# View top rows
print(df.head())
```

```python
# Tag distribution
tag_counts = df["Tag"].value_counts()
tag_counts.plot(kind='barh', title="Tag Distribution")
plt.xlabel("Count")
plt.ylabel("POS Tag")
plt.tight_layout()
plt.show()
```

```python
# Unique words count
print(f"Unique words: {df['Word'].nunique()}")
print(f"Unique tags: {df['Tag'].nunique()}")
```
